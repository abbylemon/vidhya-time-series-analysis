```python
train=pd.read_csv("Train_SU63ISt.csv") 
test=pd.read_csv("Test_0qrQsBZ.csv")
train_original=train.copy() 
test_original=test.copy()
```

```python
# make all values in the Datetime column the same format for DataFrames test and test_original

for i in range(len(test.Datetime)):
    test.Datetime.iloc[i] = str(test.Datetime.iloc[i]).replace("/", "-")

for i in range(len(test_original.Datetime)):
    test_original.Datetime.iloc[i] = str(test_original.Datetime.iloc[i]).replace("/", "-")
```

```python
# We will extract the time and date from the Datetime

train['Datetime'] = pd.to_datetime(train.Datetime,format='%d-%m-%Y %H:%M')
test['Datetime'] = pd.to_datetime(test.Datetime,format='%d-%m-%Y %H:%M')
test_original['Datetime'] = pd.to_datetime(test_original.Datetime,format='%d-%m-%Y %H:%M') 
train_original['Datetime'] = pd.to_datetime(train_original.Datetime,format='%d-%m-%Y %H:%M')
```

```python
# We made some hypothesis for the effect of hour, day, month and year on the passenger count. 
# So, let’s extract the year, month, day and hour from the Datetime to validate our hypothesis.

for i in (train, test, test_original, train_original):
    i['year']=i.Datetime.dt.year 
    i['month']=i.Datetime.dt.month 
    i['day']=i.Datetime.dt.day
    i['Hour']=i.Datetime.dt.hour 
    
train
```

```python
#  extract the day of week from Datetime and then based on the values we will assign whether the day is a weekend or not.

train['day of week']=train['Datetime'].dt.dayofweek 
temp = train['Datetime']

train
```

```python
# Let’s assign 1 if the day of week is a weekend and 0 if the day of week in not a weekend.

def applyer(row):
    if row.dayofweek == 5 or row.dayofweek == 6:
        return 1
    else:
        return 0 
temp2 = train['Datetime'].apply(applyer) 
train['weekend']=temp2

train
```

```python
# plot the training data to get an understanding for what we are working with, my plot

plt.figure(figsize = (30,10))

plt.plot_date(train_original["Datetime"], train_original["Count"])

plt.title("Time Series", fontsize = 30)
plt.xlabel("Date and Time", fontsize = 20)
plt.ylabel("Passenger Count", fontsize = 20)
plt.tight_layout()
plt.show()
```

### Lets recall the hypothesis that we made earlier:
1. Traffic will increase as the years pass by
2. Traffic will be high from May to October
3. Traffic on weekdays will be more
4. Traffic during the peak hours will be high
### After having a look at the dataset, we will now try to validate our hypothesis and make other inferences from the dataset.

```python
# Our first hypothesis was traffic will increase as the years pass by. So let’s look at yearly passenger count.

train.groupby('year')['Count'].mean().plot.bar()
```

```python
# Our second hypothesis was about increase in traffic from May to October. So, let’s see the relation between count and month.

train.groupby('month')['Count'].mean().plot.bar()
```

```python
# ere we see a decrease in the mean of passenger count in last three months. 
# This does not look right. Let’s look at the monthly mean of each year separately.

temp = train.groupby(['year', 'month'])['Count'].mean() 
temp.plot(figsize=(15,5), title= 'Passenger Count(Monthwise)', fontsize=14)
```

```python
# Let’s look at the daily mean of passenger count.

train.groupby('day')['Count'].mean().plot.bar()
```

```python
# We also made a hypothesis that the traffic will be more during peak hours. 
# So let’s see the mean of hourly passenger count.

train.groupby('Hour')['Count'].mean().plot.bar()
```

```python
# Let’s try to validate our hypothesis in which we assumed that the traffic will be more on weekdays.

train.groupby('weekend')['Count'].mean().plot.bar()
```

```python
# Now we will try to look at the day wise passenger count.
# Note - 0 is the starting of the week, i.e., 0 is Monday and 6 is Sunday.

train.groupby('day of week')['Count'].mean().plot.bar()
```

```python
"""
we will aggregate the hourly time series to daily, weekly, and monthly 
time series to reduce the noise and make it more stable and hence would 
be easier for a model to learn.
"""

train=train.drop('ID',1)

train.Timestamp = pd.to_datetime(train.Datetime,format='%d-%m-%Y %H:%M') 
train.index = train.Timestamp 
# Hourly time series 
hourly = train.resample('H').mean() 
# Converting to daily mean 
daily = train.resample('D').mean() 
# Converting to weekly mean 
weekly = train.resample('W').mean() 
# Converting to monthly mean 
monthly = train.resample('M').mean()
```

```python
# Let’s look at the hourly, daily, weekly and monthly time series.

fig, axs = plt.subplots(4,1) 
hourly.Count.plot(figsize=(15,8), title= 'Hourly', fontsize=14, ax=axs[0]) 
daily.Count.plot(figsize=(15,8), title= 'Daily', fontsize=14, ax=axs[1]) 
weekly.Count.plot(figsize=(15,8), title= 'Weekly', fontsize=14, ax=axs[2]) 
monthly.Count.plot(figsize=(15,8), title= 'Monthly', fontsize=14, ax=axs[3]) 

plt.tight_layout()
plt.show()
```
